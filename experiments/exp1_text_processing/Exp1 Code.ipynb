#9. Result Type Text Analysis
Analyze the result column:

Identify different textual result types

Count their occurrences

Visualize the distribution using a count plot

```python
# Identify different textual result types and count their occurrences
result_counts = df['result'].value_counts()
print("Result Types and their Occurrences:\n", result_counts)

# Visualize the distribution using a count plot
plt.figure(figsize=(8, 6))
sns.countplot(x='result', data=df, palette='viridis', legend=False)
plt.title('Distribution of Match Results')
plt.xlabel('Result Type')
plt.ylabel('Number of Matches')
plt.show()
```

#10. Toss Winner vs Match Winner (Text Matching)
Compare text values in toss_winner and winner:

Create a boolean column indicating whether the toss winner also won the match

Visualize the comparison using a bar chart

```python
# Create a boolean column indicating whether the toss winner also won the match
df['toss_winner_match_winner'] = df['toss_winner'] == df['winner']

# Visualize the comparison using a bar chart
plt.figure(figsize=(8, 6))
sns.countplot(x='toss_winner_match_winner', data=df, palette='viridis', legend=False)
plt.title('Toss Winner Also Won the Match?')
plt.xlabel('Toss Winner Won Match')
plt.ylabel('Number of Matches')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```
